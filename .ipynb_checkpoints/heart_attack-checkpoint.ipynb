#### Dataset

Dataset berisi mengenai data pasien yang memiliki penyakit jantung

#### Sumber dataset

data set asli berasal dari : https://www.kaggle.com/nareshbhat/health-care-data-set-on-heart-attack-possibility
        
#### objective

melakukan klasifikasi untuk mendeteksi suatu pasien berpeluang terkena serangan jantung 

##### Attribute Information

1) age
2) sex
3) chest pain type (4 values)
4) resting blood pressure
5) serum cholestoral in mg/dl
6)fasting blood sugar > 120 mg/dl
7) resting electrocardiographic results (values 0,1,2)
8) maximum heart rate achieved
9) exercise induced angina
10) oldpeak = ST depression induced by exercise relative to rest
11)the slope of the peak exercise ST segment
12) number of major vessels (0-3) colored by flourosopy
13) thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
14) target: 0= less chance of heart attack 1= more chance of heart attack

# Data Preprocessing

#### menentukan variabel independent (X) dan variabel dependent (y)

```python
X = data.drop('target', axis=1)
y = data['target']
```

#### melakukan split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

membagi data menjadi data training sebesar 0.7 dan data test sebesar 0.3

```python
X_train.shape[0]
```

data training sebesar 212 rows

```python
X_test.shape[0]
```

data test sebesar 91 rows

```python
X.shape[0]
```

# Pendefinisian dan Pelatihan Model

#### Pendefinisian dan pelatihan model Decesion Tree Classifier

```python
tree_model = DecisionTreeClassifier(min_samples_leaf=16, random_state=42)

%time tree_model.fit(X_train, y_train)
```

melakukan pendefisian model DecisionTreeClassifier dengan parameter number of samples (min_samples_leaf) sebesar 16

#### Pendefinisian dan pelatihan model Random Forest Classifier

```python
rfc_model = RandomForestClassifier(n_estimators=200, random_state=42)

%time rfc_model.fit(X_train, y_train)
```

melakukan pendefisian model RandomForestClassifier dengan parameter number of trees (n_estimators) sebesar 200

# Model Evaluation

#### Evaluasi model Decesion Tree Classifier

```python
%time y_pred_tree = tree_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_tree))
```

```python
print(classification_report(y_test, y_pred_tree))
```

Dari evaluasi model Decesion Tree Classifier diperoleh nilai accuracy 0.78

```python
y_pred_proba_tree = tree_model.predict_proba(X_test)
```

```python
tree_accuracy = accuracy_score(y_test, y_pred_tree)
tree_recall = recall_score(y_test, y_pred_tree)
tree_precision = precision_score(y_test, y_pred_tree)
tree_roc_auc_score = roc_auc_score(y_test, y_pred_proba_tree[:,1])
```

#### Evaluasi model Random Forest Classifier

```python
%time y_pred_rfc = rfc_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_rfc))
```

```python
print(classification_report(y_test, y_pred_rfc))
```

Dari evaluasi model Random Forest Classifier diperoleh nilai accuracy 0.80

```python
y_pred_proba_rfc = rfc_model.predict_proba(X_test)
```

```python
rfc_accuracy = accuracy_score(y_test, y_pred_rfc)
rfc_recall = recall_score(y_test, y_pred_rfc)
rfc_precision = precision_score(y_test, y_pred_rfc)
rfc_roc_auc_score = roc_auc_score(y_test, y_pred_proba_rfc[:,1])
```

## Model Analysis

```python
model = pd.Series(['Decesion Tree Classifier', 'Random Forest Classifier'])
```

```python
accuracy = pd.Series([tree_accuracy, rfc_accuracy])
recall = pd.Series([tree_recall, rfc_recall])
precision = pd.Series([tree_precision, rfc_precision])
roc_auc_score = pd.Series([tree_roc_auc_score, rfc_roc_auc_score])
```

```python
choose_model = pd.DataFrame({'model': model, 'accuracy': accuracy, 'recall': recall, 'precision': precision, 'roc_auc_score': roc_auc_score })
choose_model
```

dari hasil analisa perbandingan antar model Decision Tree classfier dan Random Forest Classifier terlihat bahwa model **Random Forest Classifier** memiliki nilai performa yang lebih baik yaitu dengan nilai **accuracy** 0.80, nilai **recall** 0.82, nilai **precision** 0.82, dan **roc auc score** sebesar 0.91 

## Hyperparameter Tunning

#### Melakukan hyperparameter tunning decision tree classifier

```python
# Decision Tree Classifier
tree_model = DecisionTreeClassifier(min_samples_leaf=16, random_state=42)

# hyperparameter
tree_params_grid = dict(
    criterion= ['gini', 'entropy'],
    min_samples_leaf= [16, 20]
)

# search
tree_grid = GridSearchCV(tree_model, tree_params_grid, scoring='accuracy')

%time search_tree_grid = tree_grid.fit(X,y)
```

```python
search_tree_grid.best_params_
```

```python
results = pd.DataFrame(search_tree_grid.cv_results_)[['params', 'mean_test_score']]
results
```

melakukan Hyperparameter tunning Decision Tree Classifier dengan parameter :

* criterion= ['gini', 'entropy'],
* min_samples_leaf= [16, 20]

diperoleh best parameter sebagai berikut:

* criterion= 'gini', 
* min_samples_leaf= 16 

dengan nilai accuracy **0.82**

#### Melakukan hyperparameter tunning random forest classifier

```python
# Random Forest Classifier
rfc_model = RandomForestClassifier(n_estimators=200, random_state=42)

# hyperparameter
rfc_params_grid = dict(
    n_estimators=[100,200],
    max_depth= [1, 2]
)

# search
rfc_grid = GridSearchCV(rfc_model, rfc_params_grid, scoring='accuracy')

%time search_rfc_grid = rfc_grid.fit(X,y)
```

```python
search_rfc_grid.best_params_
```

```python
results = pd.DataFrame(search_rfc_grid.cv_results_)[['params', 'mean_test_score']]
results
```

melakukan Hyperparameter tunning Random Forest Classifier dengan parameter :

* n_estimators=[100,200],
* max_depth= [1, 2]

diperoleh best parameter sebagai berikut:

* n_estimators= 200,
* max_depth= 2

dengan nilai accuracy **0.84**

# Model Inference

Mampu Mencoba model yang telah dibuat dengan random data dari masing-masing student

#### Membuat data baru 

```python
values = [[42, 0, 3 , 148, 244, 0, 0 , 178, 0 , 0.8 , 2 , 2 , 2 ]]
```

```python
data_dummies = pd.DataFrame(values,columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal',])
```

#### Melihat data baru

```python
data_dummies
```

#### melakukan prediksi dengan data baru

```python
rfc_model = RandomForestClassifier(n_estimators=200, random_state=42)
rfc_model.fit(X_train, y_train)

prediction = rfc_model.predict(data_dummies)

print("Hasil prediksi: \n")
if prediction == 0:
    print("Pasien memiliki peluang kecil terkena serangan jantung")
else:
    print("Pasien memiliki peluang besar terkena serangan jantung")
```
